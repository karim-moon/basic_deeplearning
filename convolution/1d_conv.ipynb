```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-10, 10, 1000)
y = x**2

conv_filter = np.zeros(len(x))
conv_filter[450: 550] = 1
```

```python
plt.plot(x,y)
plt.plot(x, conv_filter)
```

```python
plt.plot(np.convolve(y, conv_filter)/len(np.convolve(y, conv_filter)))
```
