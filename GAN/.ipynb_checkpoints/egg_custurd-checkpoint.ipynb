```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

from tensorflow.keras.layers import Activation, BatchNormalization, Dense, Dropout, Flatten, Reshape
from tensorflow.keras.layers import LeakyReLU
from tensorflow.keras.layers import Conv2D, Conv2DTranspose
from tensorflow.keras.models import Sequential
from tensorflow.keras.optimizers import Adam
```

```python
import os
import cv2
img_list = os.listdir('egg')

img = []
for image in img_list:
  sample_image = cv2.imread(f"egg/{image}")
  if type(sample_image) == none:
    continue
  sample_image = cv2.resize(sample_image, (200,200))
  print('image_name: ',image,'image_size: ',sample_image.shape, 'image_type: ', type(sample_image))
  img.append(sample_image)

# type(none)
```

```python
img = np.array(img)/127.5 - 1
```

```python
img_row = 200
img_col = 200
img_channel = 3

img_shape = (img_col, img_row, img_channel)
z_dim = 100
```

```python
plt.imshow(img[3],cmap='jet')
```

```python
def build_generator(z_dim):
  model = Sequential()
  model.add(Dense(100 * 25 * 25, input_dim = z_dim))
  model.add(Reshape((25,25,100)))

  model.add(Conv2DTranspose(50, kernel_size= 3, strides= 2, padding= 'same'))
  model.add(BatchNormalization())
  model.add(LeakyReLU(alpha= 0.01))

  model.add(Conv2DTranspose(25, kernel_size= 3,strides=1 ,padding= 'same'))
  model.add(BatchNormalization())
  model.add(LeakyReLU(alpha= 0.01))

  model.add(Conv2DTranspose(12, kernel_size= 3,strides=2 ,padding= 'same'))
  model.add(BatchNormalization())
  model.add(LeakyReLU(alpha= 0.01))

  model.add(Conv2DTranspose(3, kernel_size= 3, strides= 2, padding='same'))
  model.add(Activation('tanh'))

  return model
```

```python
def build_discriminator(img_shape):
  model = Sequential()

  model.add(Conv2D(32,
                   kernel_size = 3,
                   strides = 2,
                   input_shape = img_shape,
                   padding = 'same'))
  
  model.add(LeakyReLU(alpha= 0.01))

  model.add(Conv2D(64,
                   kernel_size = 3,
                   strides = 2,
                   padding = 'same'))
  
  model.add(LeakyReLU(alpha= 0.01))

  model.add(Conv2D(128,
                  kernel_size = 3,
                  strides = 2,
                  padding = 'same'))
  
  model.add(LeakyReLU(alpha= 0.01))

  model.add(Flatten())
  model.add(Dense(1, activation= 'sigmoid'))

  return model
```

```python
def build_gan(generator, discriminator):
  model = Sequential()

  model.add(generator)
  model.add(discriminator)

  return model


discriminator = build_discriminator(img_shape)
generator = build_generator(z_dim)

discriminator.compile(loss = 'binary_crossentropy',
                      optimizer = Adam(),
                      metrics = ['accuracy'])
discriminator.trainable = False

gan = build_gan(generator, discriminator)
gan.compile(loss = 'binary_crossentropy', optimizer = Adam())
```

```python
losses = []
accuracies = []
iteration_checkpoint = []

def train(iterations, batch_size, sample_interval):
 
  real = np.ones((batch_size, 1))
  fake = np.zeros((batch_size , 1))

  for iteration in range(iterations):
    
    idx = np.random.randint(0, img.shape[0], batch_size)
    imgs = img[idx]

    z = np.random.normal(0, 1, (batch_size, 100))
    gen_imgs = generator.predict(z)

    d_loss_real = discriminator.train_on_batch(imgs, real)
    d_loss_fake = discriminator.train_on_batch(gen_imgs, fake)
    d_loss, accuracy = 0.5 * np.add(d_loss_real, d_loss_fake)

    z = np.random.normal(0, 1, (batch_size, 100))
    gen_imgs = generator.predict(z)

    g_loss = gan.train_on_batch(z, real)
    if (iteration + 1) % sample_interval == 0:
      losses.append((d_loss, g_loss))
      accuracies.append(100.0 * accuracy)
      iteration_checkpoint.append(iteration + 1)

      print(f'iter : {iteration + 1}[D손실 : {d_loss}], [정확도 : {100.0 * accuracy}] [G손실 : {g_loss}]')

      sample_images(generator)
```

```python
def sample_images(generator, image_grid_rows = 2, image_gird_columns = 2):
  z = np.random.normal(0, 1, (image_grid_rows * image_gird_columns, z_dim))

  gen_imgs = generator.predict(z)
  gen_imgs = 0.5 * gen_imgs + 0.5

  fig, axs = plt.subplots(image_grid_rows,
                          image_gird_columns,
                          figsize = (4,4),
                          sharex = True,
                          sharey = True)
  
  cnt = 0 
  for i in range(image_grid_rows):
    for j in range(image_gird_columns):
      axs[i, j].imshow(gen_imgs[cnt, :, :])
      axs[i, j].axis('off')
      cnt += 1

  plt.show()
```

```python
iterations = 10000
batch_size = 30
sample_interval = 1000

train(iterations, batch_size= batch_size , sample_interval= sample_interval)
```
